```python
from rdkit import Chem
from rdkit.Chem import AllChem, Draw
from rdkit.Chem.Draw import IPythonConsole
IPythonConsole.drawOptions.addAtomIndices = True
IPythonConsole.molSize = (400,400)
import tabs
```

```python
m1 = Chem.AddHs(Chem.MolFromSmiles("CCCCO"))
m1
```

```python
tabs.GetMultiplicityAllBonds(m1)
```

```python
m2 = Chem.AddHs(Chem.MolFromSmiles("C(CO)CC"))
m2
```

```python
tabs.GetMultiplicityAllBonds(m2)
```

```python
matching = m2.GetSubstructMatch(m1)
```

```python
m2matched =  Chem.RenumberAtoms(m2, matching)
```
